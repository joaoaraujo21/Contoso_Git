```python
%sql
USE CATALOG workspace;
USE SCHEMA datamart;
```

```python
# Configurando caminho
base_path = "/Volumes/workspace/datamart/varejo/"

base_customers = f"{base_path}customers.csv"
base_orders = f"{base_path}orders.json"
base_payments = f"{base_path}payments.csv"
base_products = f"{base_path}products.parquet"
base_reviews = f"{base_path}reviews.json"
```

```python
# Transformando em Dataframe

df_customers = spark.read.format("csv").option("header", True).option("inferSchema", True).option("sep", ",").load(base_customers)
df_payments = spark.read.format("csv").option("header", True).option("inferSchema", True).option("sep", ",").load(base_payments)
df_orders = spark.read.format("json").option("multiline", True).load(base_orders)
df_reviews = spark.read.format("json").option("multiline", True).load(base_reviews)
df_products = spark.read.parquet(base_products)

#display(df_products)
```

```python
# Salvar tabelas na camada Bronze
#.write.format("delta").mode("overwrite").saveAsTable("bronze_sales")

df_customers.write.format("delta").mode("overwrite").saveAsTable("workspace.datamart.bronze_customers")
df_orders.write.format("delta").mode("overwrite").saveAsTable("workspace.datamart.bronze_orders")
df_payments.write.format("delta").mode("overwrite").saveAsTable("workspace.datamart.bronze_payments")
df_reviews.write.format("delta").mode("overwrite").saveAsTable("workspace.datamart.bronze_reviews")
df_products.write.format("delta").mode("overwrite").saveAsTable("workspace.datamart.bronze_products")

print("âœ… Sucesso! Dados ingeridos e tabelas Bronze criadas.")
```
